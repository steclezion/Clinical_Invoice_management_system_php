```python
from pyspark.sql import SparkSession
import re
```

```python
# Create a SparkSession
spark_master = SparkSession.builder \
    .master("local[*]") \
    .appName("SixWordSequenceCount") \
    .getOrCreate()
```

```python
# Get the SparkContext from the SparkSession
sc = spark_master.sparkContext
```

```python
# Read the input file (adjust the path to where W25-Sample.txt is located)
Readlines = sc.textFile("W25-Sample.txt")
```

```python
def six_word_sequences_extract(line):
    """
    Given a line of text, remove punctuation, split into words, and return
    all contiguous six-word sequences.
    The reason I am removing punctuation is due the question is clearly stating only words of sequence interms of six count
    """

    # Remove punctuation (you can adjust the regex if needed)
    cleaned_line = re.sub(r'[^\w\s]', '', line)

    # Split into words and convert to lower case (optional normalization)
    words = cleaned_line.lower().split()

    # If you do not want to convert into lowercase (that us )
    words = cleaned_line.split()

    sequences = []
    # Only process if there are at least 6 words in this line.

    if len(words) >= 6:
        for i in range(len(words) - 5):
            # Join 6 consecutive words into a single string.
            seq = " ".join(words[i:i+6])
            sequences.append(seq)

    return sequences
```

```python
# Use flatMap to convert each line into 6-word sequences.
# flatMap is applied to each element in the readlines and helps to flatten the result
six_word_sequences = Readlines.flatMap(six_word_sequences_extract)
```

```python
# Map each sequence to a (sequence, 1) pair.
# Map helps to map to each element of an RDD(Resilient Distributed DataSet) and creates a new element with transformed element
sequence_pairs = six_word_sequences.map(lambda seq: (seq, 1))
```

```python
# Reduce by key to count the occurrences of each sequence.
sequence_counts = sequence_pairs.reduceByKey(lambda a, b: a + b)
```

```python
# Filter to only include sequences with a count greater than 1
filtered_sequence_counts = sequence_counts.filter(lambda x: x[1] > 1)

sequence_counts = sequence_counts.sortBy(lambda x: x[1], ascending=True)
```

```python
# Collect and display the sorted results
results = sequence_counts.collect()
#results = filtered_sequence_counts.collect()

for sequence, count in results:
    print(f"{sequence}: {count}")
```
